# Sistema Financeiro Familiar com SQLAlchemy
Este notebook demonstra como utilizar as classes ORM para manipular o banco de dados MySQL com as tabelas `categoria`, `conta`, `usuarios` e `financas`.

```python
# Importar as classes e funções definidas
from sistema_financeiro_familiar_completo import (
    adicionar_categoria, listar_categorias, atualizar_categoria, remover_categoria,
    adicionar_conta, listar_contas, atualizar_conta, remover_conta,
    adicionar_usuario, listar_usuarios, atualizar_usuario, remover_usuario,
    adicionar_financa, listar_financas, atualizar_financa, remover_financa
)
```

## Manipulação de Categorias

```python
# Adicionar uma nova categoria
nova_categoria = adicionar_categoria('Alimentação', 'Despesas com alimentos')
print(f'Categoria adicionada: {nova_categoria.desc_categoria}')

# Listar categorias
categorias = listar_categorias()
for categoria in categorias:
    print(f'{categoria.idcategoria}: {categoria.desc_categoria} - {categoria.obs_categoria}')
```

## Manipulação de Contas

```python
# Adicionar uma nova conta
nova_conta = adicionar_conta('Conta Corrente', 'Principal conta bancária')
print(f'Conta adicionada: {nova_conta.conta}')

# Listar contas
contas = listar_contas()
for conta in contas:
    print(f'{conta.idconta}: {conta.conta} - {conta.desc_conta}')
```

## Manipulação de Usuários

```python
# Adicionar um novo usuário
novo_usuario = adicionar_usuario('João Silva', 'joao@example.com')
print(f'Usuário adicionado: {novo_usuario.nome_usuario}')

# Listar usuários
usuarios = listar_usuarios()
for usuario in usuarios:
    print(f'{usuario.idusuarios}: {usuario.nome_usuario} - {usuario.email_usuario}')
```

## Manipulação de Finanças

```python
# Adicionar uma nova transação financeira
nova_financa = adicionar_financa(1, 1, 1, 'Compra no mercado', 200)
print(f'Transação financeira adicionada: {nova_financa.descricao} - {nova_financa.valor}')

# Listar transações financeiras
financas = listar_financas()
for financa in financas:
    print(f'{financa.idfinancas}: {financa.descricao} - {financa.valor}')
```
